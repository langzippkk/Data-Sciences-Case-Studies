```python
train = pd.read_csv('..\\0.data\\trainV1.csv')
test = pd.read_csv('..\\0.data\\testV1.csv')

print(train.shape)
print(test.shape)
```

### Training Models

```python
trainY = train.loc[:, ['Survived']]
trainX = train.iloc[:, 2:]

testX = test.iloc[:, 1:]
```

```python
print(train.shape)
print(test.shape)

print(trainY.shape)
print(testX.shape)
```

```python
trainX
```

## Linear Regression

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
#Create linear regression object
linear = linear_model.LinearRegression()

#Train the model using the training sets and
#check score
linear.fit(trainX, trainY)
train_r2Score = linear.score(trainX, trainY)

#Predict Output
y_predicted = linear.predict(testX)

print('Intercept: ', linear.intercept_)
print('\nTraining R2 Score: ', train_r2Score)
```
